```python
import pandas as pd
import numpy as np

# Cargo el Dataset
df_hotel = pd.read_csv('/workspaces/LucasDellaMaestra-AlejandroNovellino-ReneRamirez-ProyectoFinal/data/raw/hotel_booking.csv')

# Mostramos para chequear
print(df_hotel.head())
```

```python
# Eliminamos columnas que no aportan
columnas_a_eliminar = ['name', 'email', 'phone-number', 'credit_card', 'company', 'reservation_status', 'reservation_status_date']
df_hotel.drop(columns=columnas_a_eliminar, inplace=True)

print("\nColumnas luego de eliminar las que no sirven:")
print(df_hotel.columns)

# Ahora analizamos los valores nulos
print("\nCantidad de valores nulos por columna:")
print(df_hotel.isnull().sum())

# Para ver también en porcentaje (opcional)
print("\nPorcentaje de valores nulos por columna:")
print((df_hotel.isnull().sum() / len(df_hotel)) * 100)
```

#### Graficos De Análisis

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Aseguramos estilo copado
sns.set(style="whitegrid")

# 1. hotel vs. cancelación
plt.figure(figsize=(8,5))
sns.countplot(data=df_hotel, x='hotel', hue='is_canceled')
plt.title('Hotel vs Cancelación')
plt.xlabel('Tipo de Hotel')
plt.ylabel('Cantidad de Reservas')
plt.legend(title='Cancelada', labels=['No', 'Sí'])
plt.show()

# 2. lead_time vs. cancelación (histograma)
plt.figure(figsize=(8,5))
sns.histplot(data=df_hotel, x='lead_time', hue='is_canceled', kde=True, element="step", stat="density", common_norm=False)
plt.title('Lead Time vs Cancelación')
plt.xlabel('Lead Time (días)')
plt.ylabel('Densidad')
plt.legend(title='Cancelada', labels=['No', 'Sí'])
plt.show()

# 3. market_segment vs. cancelación
plt.figure(figsize=(10,5))
sns.countplot(data=df_hotel, x='market_segment', hue='is_canceled')
plt.title('Segmento de Mercado vs Cancelación')
plt.xlabel('Segmento de Mercado')
plt.ylabel('Cantidad de Reservas')
plt.legend(title='Cancelada', labels=['No', 'Sí'])
plt.xticks(rotation=45)
plt.show()

# 4. adr vs. cancelación (histograma)
plt.figure(figsize=(8,5))
sns.histplot(data=df_hotel, x='adr', hue='is_canceled', kde=True, element="step", stat="density", common_norm=False)
plt.title('ADR (Average Daily Rate) vs Cancelación')
plt.xlabel('ADR (precio medio por noche)')
plt.ylabel('Densidad')
plt.xlim(0, 500)  # Limitar para que los outliers extremos no deformen el gráfico
plt.legend(title='Cancelada', labels=['No', 'Sí'])
plt.show()
```
